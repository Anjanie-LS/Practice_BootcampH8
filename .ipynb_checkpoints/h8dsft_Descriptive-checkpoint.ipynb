- Cari Measure of Central Tendency dari data tersebut.
- Cari Measure of Variance dari data tersebut.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')
print(df)
```

Finding the mean:

```python
print (df['New Cases'].mean())
print (df['New Deaths'].mean())
print (df['New Recovered'].mean())
print (df['New Active Cases'].mean())
print (df['Total Cases'].mean())
print (df['Total Deaths'].mean())
```

Finding the median:

```python
df.sort_values('New Cases')
df.sort_values('New Deaths')
df.sort_values('New Recovered')
df.sort_values('New Active Cases')
df.sort_values('Total Cases')
df.sort_values('Total Deaths')

df
```

```python
print (df['New Cases'].median())
print (df['New Deaths'].median())
print (df['New Recovered'].median())
print (df['New Active Cases'].median())
print (df['Total Cases'].median())
print (df['Total Deaths'].median())
```

Finding the mode:

```python
print (df['New Cases'].mode())
print (df['New Deaths'].mode())
print (df['New Recovered'].mode())
print (df['New Active Cases'].mode())
print (df['Total Cases'].mode())
print (df['Total Deaths'].mode())
```

```python
numcols = ['New Cases', 'New Deaths', 'New Recovered', 'New Active Cases' ,'Total Cases', 'Total Deaths']
for col in numcols:
    print(df[col].name + ' range: ' + str(df[col].max() - df[col].min()))
```

```python
print (df['New Cases'].var())
print (df['New Deaths'].var())
print (df['New Recovered'].var())
print (df['New Active Cases'].var())
print (df['Total Cases'].var())
print (df['Total Deaths'].var())
```
